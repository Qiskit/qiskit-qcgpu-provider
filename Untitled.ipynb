```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import math
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit import execute, Aer
from qiskit.tools.visualization import circuit_drawer
from qiskit.backends.ibmq import least_busy
import qiskit_addon_qcgpu


plt.style.use('ggplot')

def plot(data_qiskit, data_qcgpu):
    state_qiskit = np.power(np.abs(data_qiskit), 2)
    state_qcgpu = np.power(np.abs(data_qcgpu), 2)

    ind = np.arange(state_qiskit.size) 
    width = 0.35
    plt.bar(ind, state_qiskit, width, label='qiskit')
    plt.bar(ind + width, state_qcgpu, width,
        label='qcpu')

    plt.ylabel('Outcome Probability')
        
    plt.xticks(ind + width / 2, ind)
    plt.legend(loc='best')
    plt.show()
    
def compare(qc):
    sim_backend = qiskit_addon_qcgpu.get_backend('local_statevector_simulator_qcgpu')
    job = execute(qc, sim_backend, shots=1)
    result1 = job.result()

    data1 = result1.get_data()['statevector']
    
    sim_backend = Aer.get_backend('statevector_simulator')
    job = execute(qc, sim_backend, shots=1)
    result2 = job.result()

    data2 = result2.get_data()['statevector']
    
    plot(data2, data1)
    
    
```

```python
q = QuantumRegister(3, "q")
qc = QuantumCircuit(q)

qc.h(q[0])
qc.h(q[2])
qc.cx(q[0], q[1])
qc.cx(q[2], q[0])

circuit_drawer(qc)
```

```python
compare(qc)
```
